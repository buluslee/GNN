## 导入数据集ACM

```python
from scipy import io
mat_file = io.loadmat('ACM.mat')
print(mat_file.keys())
print(mat_file)
```

## 1. 获取paper节点的类型标签

## 获取所有的paper的所有的类型

```python
##获取有paper_conf的边连接的paper
paper_conf = mat_file['PvsC'].nonzero()[1]
```

## 提取paper类型为 database的paper索引

```python
import numpy as np
# DataBase

##建立bool数组判断每个文章是否属于DataBase类别
paper_db = np.isin(paper_conf,[1,13])

##检查‘paper_conf’中的元素是否在[1,13]这个列表中，如果是就返回True,反之则返回'False'
paper_db_idx = np.where(paper_db == True)[0]

## 获取DataBase类别的论文索引，将这些论文存储在paper_db_idx中
paper_db_idx = np.sort(np.random.choice(paper_db_idx,994,replace=False))

print(paper_db_idx)
```

## 提取paper类型为 Data Mining的paper索引

```python
# Data Mining

## 检查 paper_conf 中的元素是否在 [0] 这个列表中，如果是，就在相应位置返回 True，否则返回 False
paper_dm = np.isin(paper_conf,[0])

##获取所有属于 "Data Mining" 类别的论文的索引，这些索引存储在 paper_dm_idx 中
paper_dm_idx = np.where(paper_dm == True)[0]
print(paper_dm_idx)
```

## 提取paper类型为 Wireless Communication 的paper索引

```python
# Wireless Communication

##检查 paper_conf 中的元素是否在 [9, 10] 这个列表中，如果是，就在相应位置返回 True，否则返回 False
paper_wc = np.isin(paper_conf,[9,10])

##通过使用 np.where(paper_wc == True) 获取所有属于 "Wireless Communication" 类别的论文的索引，这些索引存储在 paper_wc_idx 中
paper_wc_idx = np.where(paper_wc == True)[0]
print(paper_wc_idx)
```

## 得到三种paper类型即database，datamining，Wireless Communication的总集合——即我们需要分析的paper

```python
## 得到全部的paper索引并按照从小到大排列
paper_idx = np.sort(list(paper_db_idx)+list(paper_dm_idx)+list(paper_wc_idx))
print(paper_idx)
len(paper_idx)
```

## 给这三种paper分类进行重构索引

```python
paper_target = []
## 遍历索引
for idx in paper_idx:
    ## paper 类型为database的索引记为0
    if idx in paper_db_idx:
        paper_target.append(0)
    ## paper 类型为datamining的索引记为1
    elif idx in paper_wc_idx:
        paper_target.append(1)
    ## paper 类型为 Wireless Communication的索引记为2
    else:
        paper_target.append(2)
## 转化为数组
paper_target = np.array(paper_target)
print(paper_target.shape)
```

## 提炼author并对其编码

```python
## 选出所涉及的三种类型的论文（[paper_idx]）在PvsA中对应的Author
authors = mat_file['PvsA'][paper_idx].nonzero()[1]

##创建authort字典将原始索引和新索引的映射关系
author_dic = {}

##创建author新索引的列表
re_authors = []

##遍历author的原始索引
for author in authors:
    if author not in author_dic:
        ##key为在author新索引（paper索引后进行author新索引生成），value为原始索引
        author_dic[author] = len(author_dic) + len(paper_idx)
    ##将key即author新索引添加到列表即生成re_authors
    re_authors.append(author_dic[author])
re_authors = np.array(re_authors)

print(f'文章数量:{len(paper_idx)}篇')
print(f'作者数量:{len(author_dic)}位')
```

## 提炼subject并对其编码

```python
## 选出所涉及的三种类型的论文（[paper_idx]）在PvsL中对应的subject
subjects = mat_file['PvsL'][paper_idx].nonzero()[1]

##创建subject字典将原始索引和新索引的映射关系
subject_dic = {}

##创建subject新索引的列表
re_subjects = []

##遍历subject的原始索引
for subject in subjects:
    if subject not in subject_dic:
    ##key为在subject新索引（auhthor索引后进行subject新索引生成），value为原始索引
        subject_dic[subject] = len(subject_dic) + len(paper_idx) + len(author_dic)
    ##将key即subject新索引添加到列表即生成re_subject
    re_subjects.append(subject_dic[subject])
re_subjects = np.array(re_subjects)

print(f'主题数量:{len(subject_dic)}个')
print(re_subjects)
```

## 统计节点数量

```python
node_num = len(paper_idx) + len(author_dic) + len(subject_dic)
print(node_num)
```

## 构建不同边类型的边邻接矩阵

```python
from scipy.sparse import csr_matrix

## 构建paper-author的边类型的矩阵
papers_A = mat_file['PvsA'][paper_idx].nonzero()[0]
data = np.ones_like(papers_A)
A_pa = csr_matrix((data, (papers_A, re_authors)), shape=(node_num,node_num))
print(A_pa)


##  构建paper-subject的边类型的矩阵
papers_L = mat_file['PvsL'][paper_idx].nonzero()[0]
data = np.ones_like(papers_L)
A_ps = csr_matrix((data, (papers_L, re_subjects)), shape=(node_num,node_num))
print(A_ps)


## 构建author-paper的边类型的矩阵
A_ap = A_pa.transpose()

##构建subject-paper的边类型的矩阵
A_sp = A_ps.transpose()
```

## 统计边的类型

```python
edges = [A_pa,A_ap,A_ps,A_sp]
```

## 节点特征构建-引入term协助构建节点特征矩阵

```python
##选择paper_idx所选择的特定的论文索引的论文的主题
terms = mat_file['TvsP'].transpose()[paper_idx].nonzero()[1]

##创建term字典将原始索引和新索引的映射关系
term_dic = {}

##创建term新索引的列表
re_terms = []

##遍历term的原始索引
for term in terms:
    if term not in term_dic:
        ##key为在subject新索引（subject索引后进行term新索引生成），value为原始索引
        term_dic[term] = len(term_dic) + len(paper_idx) + len(author_dic) + len(subject_dic)
    ##将key即term新索引添加到列表即生成re_terms
    re_terms.append(term_dic[term])
re_terms = np.array(re_terms)

print(len(term_dic))
```

## 获取不同节点类型的节点特征矩阵

```python
## 将代表terms的节点加入之前已包含：author、paper、subject
tmp_num_node = node_num + len(term_dic)

##构建paper-author关系矩阵 A_pa_tmp
papers = mat_file['PvsA'][paper_idx].nonzero()[0]
data = np.ones_like(papers)
A_pa_tmp = csr_matrix((data, (papers, re_authors)), shape=(tmp_num_node,tmp_num_node))

## 构建paper-subject关系矩阵 A_ps_tmp
papers = mat_file['PvsL'][paper_idx].nonzero()[0]
data = np.ones_like(papers)
A_ps_tmp = csr_matrix((data, (papers, re_subjects)), shape=(tmp_num_node,tmp_num_node))

## 构建paper-terms关系矩阵 A_pt_tmp
papers = mat_file['PvsT'][paper_idx].nonzero()[0]
data = np.ones_like(papers)
A_pt_tmp = csr_matrix((data, (papers, re_terms)), shape=(tmp_num_node,tmp_num_node))
```

```python
paper_feat = np.array(A_pt_tmp[:len(paper_idx),-len(term_dic):].toarray()>0, dtype=np.int32)
author_feat = np.array(A_pa_tmp.transpose().dot(A_pt_tmp)[len(paper_idx):len(paper_idx)+len(author_dic),-len(term_dic):].toarray()>0, dtype=np.int32)
subject_feat = np.array(A_ps_tmp.transpose().dot(A_pt_tmp)[len(paper_idx)+len(author_dic):len(paper_idx)+len(author_dic)+len(subject_dic),-len(term_dic):].toarray()>0, dtype=np.int32)
print(paper_feat.shape)
print(author_feat.shape)
print(subject_feat.shape)
```

```python
##拼接节点特征向量
node_features = np.concatenate((paper_feat,author_feat,subject_feat))
print(node_features.shape)
```

## 定义GTConv



```python
from torch import nn
import torch.nn.functional as F
import math
```

```python
class GTConv(nn.Module):
    
    def __init__(self, in_channels, out_channels):
        super(GTConv, self).__init__()
        self.in_channels = in_channels
        self.out_channels = out_channels
        ## 卷积核定义其形状（out_channels,in_channels,1,1)
        self.weight = nn.Parameter(torch.Tensor(out_channels,in_channels,1,1))
        self.bias = None
        self.scale = nn.Parameter(torch.Tensor([0.1]), requires_grad=False)
        self.reset_parameters()
## 初始化权重和偏置项的数值
    def reset_parameters(self):
        n = self.in_channels
        nn.init.constant_(self.weight, 0.1)
        ##如果存在偏置项，则计算权重的扇入和扇出
        if self.bias is not None:
            fan_in, _ = nn.init._calculate_fan_in_and_fan_out(self.weight)
            bound = 1 / math.sqrt(fan_in)
        ##nn.init.uniform_函数将偏置项的值初始化为均匀分布的随机数
            nn.init.uniform_(self.bias, -bound, bound)
            
##在A的一维上进行加权求和（即对每张边类型图的重要性进行加权求和）    
    def forward(self, A):
        A = torch.sum(A*F.softmax(self.weight, dim=1), dim=1)
        return A
```

## 定义GTlayer

```python
class GTLayer(nn.Module):
    
    def __init__(self, in_channels, out_channels, first=True):
        super(GTLayer, self).__init__()
        self.in_channels = in_channels
        self.out_channels = out_channels
        self.first = first
        if self.first == True:
            self.conv1 = GTConv(in_channels, out_channels)
            self.conv2 = GTConv(in_channels, out_channels)
        else:
            self.conv1 = GTConv(in_channels, out_channels)
    
    def forward(self, A, H_=None):
    ##如果是第一层
        if self.first == True:
        ##单独进行两次卷积
            a = self.conv1(A)
            b = self.conv2(A)
        ##将经过设定了边类型权重的A进行矩阵相乘生成元路径图
            H = torch.bmm(a,b) 
        ##  
            W = [(F.softmax(self.conv1.weight, dim=1)).detach(),(F.softmax(self.conv2.weight, dim=1)).detach()]
        else:
            a = self.conv1(A)
            H = torch.bmm(H_,a)
            W = [(F.softmax(self.conv1.weight, dim=1)).detach()]
        return H,W
```

## 定义模型GTN

```python
class GTN(nn.Module):
    
    def __init__(self, num_edge, num_channels, w_in, w_out, num_class,num_layers,norm):
        super(GTN, self).__init__()
        self.num_edge = num_edge#5
        self.num_channels = num_channels#2
        self.w_in = w_in#输入节点维度1902
        self.w_out = w_out#输出节点维度
        self.num_class = num_class#节点分类
        self.num_layers = num_layers#几层gtlayer
        self.is_norm = norm#ture
        layers = []
        for i in range(num_layers):#0、1、2
            if i == 0:
                layers.append(GTLayer(num_edge, num_channels, first=True))
            else:
                layers.append(GTLayer(num_edge, num_channels, first=False))
        self.layers = nn.ModuleList(layers)
        ##设定节点特征维度变化矩阵
        self.weight = nn.Parameter(torch.Tensor(w_in, w_out))
        ##设定偏置
        self.bias = nn.Parameter(torch.Tensor(w_out))
        ##设定损失
        self.loss = nn.CrossEntropyLoss()
        ##第一个线性层的shape（输出节点维度*通道数，输出节点维度）
        self.linear1 = nn.Linear(self.w_out*self.num_channels, self.w_out)
        ##第二个线性层的shape（输出节点维度，分类数量）
        self.linear2 = nn.Linear(self.w_out, self.num_class)
        self.reset_parameters()
## 参数重置
    def reset_parameters(self):
        nn.init.xavier_uniform_(self.weight)
        nn.init.zeros_(self.bias)
## 定义gcn卷积
    def gcn_conv(self,X,H):
        ##是X节点特征矩阵（num_nodes, in_feature)
        X = torch.mm(X, self.weight)
        ##生成的元路径异构图进行标准化
        H = self.norm(H, add=True)
        return torch.mm(H.t(),X)
## 对通道中生成的元路径图H进行标准化
    def normalization(self, H):
        for i in range(self.num_channels):
            if i==0:
                H_ = self.norm(H[i,:,:]).unsqueeze(0)
            else:
                H_ = torch.cat((H_,self.norm(H[i,:,:]).unsqueeze(0)), dim=0)
        return H_
## 定义
    def norm(self, H, add=False):
        ##转置
        H = H.t()
        if add == False:
            
            H = H*((torch.eye(H.shape[0])==0).type(torch.FloatTensor))
        else:
        ##将 H 的对角线元素置零, 恢复 H 的对角线元素为 1。
            H = H*((torch.eye(H.shape[0])==0).type(torch.FloatTensor)) + torch.eye(H.shape[0]).type(torch.FloatTensor)
        ##计算 H 每行的和，表示每个节点的度（连接数）
        deg = torch.sum(H, dim=1)
        ##计算度的倒数
        deg_inv = deg.pow(-1)
        ##将无限大的倒数值设置为 0，这通常对应于没有连接的节点
        deg_inv[deg_inv == float('inf')] = 0
        ##将度的倒数转换为对角矩阵
        deg_inv = deg_inv*torch.eye(H.shape[0]).type(torch.FloatTensor)
        ##通过与度矩阵的逆相乘，归一化 H
        H = torch.mm(deg_inv,H)
        ##转置回来
        H = H.t()
        return H

    def forward(self, A, X, target_x, target):
        ##A, node_features, train_node, train_target
        A = A.unsqueeze(0).permute(0,3,1,2) 
        Ws = []
        for i in range(self.num_layers):
            if i == 0:
                H, W = self.layers[i](A)
            else:
                H = self.normalization(H)
                H, W = self.layers[i](A, H)
            Ws.append(W)
        
        #H,W1 = self.layer1(A)
        #H = self.normalization(H)
        #H,W2 = self.layer2(A, H)
        #H = self.normalization(H)
        #H,W3 = self.layer3(A, H)
        
        for i in range(self.num_channels):
            if i==0:
                X_ = F.relu(self.gcn_conv(X,H[i]))
                
            else:
                X_tmp = F.relu(self.gcn_conv(X,H[i]))
                X_ = torch.cat((X_,X_tmp), dim=1)
        X_ = self.linear1(X_)
        X_ = F.relu(X_)
        y = self.linear2(X_[target_x])
        loss = self.loss(y, target)
        return loss, y, Ws
```
